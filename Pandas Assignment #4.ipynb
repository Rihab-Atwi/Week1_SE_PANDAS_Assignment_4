```python
df1 = df.groupby('bedrooms')['price'].mean().reset_index()
```

**The data reveals that houses with more bedrooms tend to have higher average prices. For instance, houses with 1 bedroom show an average price of 353,042.86 while those with 2 bedrooms show an average price of 415,767.86 and this trend continues for higher numbers of bedrooms.**

```python
df[['bedrooms', 'sqft_living', 'sqft_lot']].groupby('bedrooms').mean().round(1).head(4)
```

**The average living space and lot size increase with the number of bedrooms. This is to be expected, as houses with more bedrooms tend to be larger and have larger lots.**

```python
df[['price', 'grade', 'sqft_living', 'sqft_lot']].groupby('grade').mean().round(1).reset_index()
```

**Houses with a higher grade tend to be more expensive, have more square footage, and have larger lot sizes.**

```python
df[['price', 'waterfront', 'sqft_living', 'sqft_lot']].groupby('waterfront').mean().round(1).reset_index()
```

**Waterfront and House Prices:**

<br>

**Living Area and Lot Area:**
- We can assume that waterfront properties might have larger living and lot areas on average.

```python
df[['price', 'waterfront', 'grade','sqft_living', 'sqft_lot']].groupby(['waterfront', 'grade']).mean().round(1)
```

- Generally, houses renovated in earlier years tend to have lower average prices.
- Houses renovated in more recent years show higher average prices.
- Houses that have been renovated more recently tend to have higher average prices, suggesting that modern updates and improvements can add value to a house.

```python
renovated_df.groupby('yr_renovated').size().reset_index()
```

- The number of houses that underwent renovation increased in recent years, possibly due economic conditions.

# Correlation

```python
correlations = df.corr()

print(correlations)
```

The output of the code is a correlation matrix, which shows the correlation between all pairs of columns in the dataset. The correlation coefficient is a number between -1 and 1, where:
- A correlation coefficient of -1 indicates a perfect negative correlation.
- A correlation coefficient of 0 indicates no correlation.
- A correlation coefficient of 1 indicates a perfect positive correlation.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("kc_house_data.csv")

correlations = df.corr()

plt.matshow(correlations, cmap="Blues")
plt.title("Correlation Matrix for KC_HOSES_DATA")
plt.show()
```
